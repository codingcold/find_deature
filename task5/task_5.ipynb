```python
#read dataset1
dataset1=pd.read_excel('3datasets/dataset1/NSF-metadata.xlsx',engine='openpyxl')
dataset1
```

```python
#extracting features: DOI,Title,Creator/Author,Journal Name,Sponsoring Org.
dataset_1_new=dataset1.loc[:,['DOI','Title','Creator/Author','Journal Name','Sponsoring Org.']]
dataset_1_new
```

```python
#read dataset2
dataset_2=pd.read_csv('3datasets/dataset2/Colleges_and_Universities.csv')
dataset_2
```

```python
#extracting features: NAME,ADDRESS,STATE,COUNTRY,COUNTY,SOURCE,WEBSITE
dataset_2_new=dataset_2.loc[:,['NAME','ADDRESS','STATE','COUNTRY','COUNTY','SOURCE','WEBSITE']]
dataset_2_new
```

```python
#read dataset3
f=open("3datasets/dataset3/QSrank.txt")
line = f.readline()
columns=[]
while line:
    columns.append(line.rstrip('\n'))
    line = f.readline()
columns
```

```python
#split 
all_data_3=[]
for i in range(len(columns)):
    chunks = re.split('  +', columns[i])
    all_data_3.append(chunks)
print(all_data_3)

```

```python
#split 
all_data_3=[]
for i in range(len(columns)):
    chunks = re.split('  +', columns[i])
    all_data_3.append(chunks)
print(all_data_3)
```

```python
#extracting features:rank,name,area
ranks=[]
names=[]
areas=[]
for data in all_data_3:
    ranks.append(data[0])
    names.append(data[1])
    areas.append(data[2])


data_3={
    'UNI_RANKS':ranks,
    'UNI_NAMES':names,
    'UNI_AREAS':areas
}

d3=pd.DataFrame(data_3,index=None)
d3
    
```

```python
#merge all datasets to the BIK
dataset_all = pd.concat([bik,dataset_1_new, dataset_2_new,d3], axis=1)
dataset_all
```
